# How to Optimize Loops in Python  | Speed up Your Python Codes | A faster way to loop using built-in functions

```python
import time
start = time.time()
total_sum = 0
for i in range(10000000):
    total_sum += i

print("Sum = ",total_sum)
print("Total Time =",time.time() - start ,"seconds")
```

```python
#Use Built-in Function 
import time
start = time.time()

total_sum = sum(range(10000000))

print("Sum = ",total_sum)
print("Total Time =",time.time() - start ,"seconds")
```

```python
#Use Numpy
import time
import numpy as np
start = time.time()
total_sum = np.sum(np.arange(10000000))

print("Sum = ",total_sum)
print("Total Time =",time.time() - start ,"seconds")
```

```python
import time
import numpy as np

random_scores = np.random.randint(1, 100, size=10000000)

start = time.time()

count_failed = 0
sum_failed = 0
for score in random_scores:
    if score < 70:
        sum_failed += score
        count_failed += 1

print(sum_failed/count_failed)
print("Total Time =",time.time() - start ,"seconds")
```

```python
#Use Numpy Built-in Functions
import time
import numpy as np

random_scores = np.random.randint(1, 100, size=100000010)

start = time.time()

mean_failed = (random_scores[random_scores < 70]).mean()
print(mean_failed)

print("Total Time =",time.time() - start ,"seconds")
```
